## 데이터 불러오기

```python
import pandas as pd
```

```python
word_count = pd.read_csv("data/word_count.csv", index_col=0)
word_count.head()
```

## 단어 구름

```python
from wordcloud import WordCloud
```

### `WordCloud`의 중요 옵션
- `font_path`: 글꼴의 경로
- `max_words`: 워드클라우드를 그릴 단어의 개수
- `background_color`: 배경색 설정
- `width`: 가로크기(픽셀 단위)
- `height`: 세로크기(픽셀 단위)

```python
# 단어 구름 설정
wc = WordCloud(background_color='white', max_words=100, width=400, height=300)
# 배경은 흰색, 최대 100단어, 가로 400픽셀, 세로 300픽셀 크기
```

### 데이터프레임을 `WordCloud`에 넣을 수 있게 변환
- `to_dict`함수를 이용해 사전(`dict`) 형태로 변환
 - 사전(`dict`): Key와 값을 짝지은 자료형

단어가 Key, 빈도가 값이 되도록 변환

```python
count_dic = word_count.set_index('단어')['빈도'].to_dict()
count_dic
```

#### 또 다른 방법

```python
dict(zip(word_count['단어'], word_count['빈도']))
```

### 시각화

```python
cloud = wc.fit_words(count_dic)
```

```python
cloud.to_image()
```

## 파일로 저장

```python
cloud.to_file('data/cloud.png')
```
